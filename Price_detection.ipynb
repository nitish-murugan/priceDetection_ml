```python
# Mini Assignment

import pandas as pd

df = pd.read_csv("gemstone.csv")

from sklearn.preprocessing import LabelEncoder

encoders = {}

for col in df.columns:
    if col in ['cut', 'color', 'clarity']:
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
        encoders[col] = encoder

print(df.head())
X = df[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z']]
Y = df['price']

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=43)

model = LinearRegression()
model = model.fit(x_train, y_train)

yPred = model.predict(x_test)

mse = mean_squared_error(y_test, yPred)
print(f"Mean Squared Error: {mse}")

r2 = r2_score(y_test, yPred)
print(f"R-squared: {r2}")


neededData = pd.DataFrame({"carat": [1.50], "cut": [3], "color": [2], "clarity": [4], "depth": [61.5], "table": [58.0], "x": [6.0], "y": [6.0], "z": [4.0]})

for col in neededData.columns:
    if col in ['cut', 'color', 'clarity']:
        neededData[col] = encoders[col].fit_transform(neededData[col])

print(f"Price for the given input is {model.predict(neededData)[0]}")
```
